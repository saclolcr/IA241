# What is the Total Death Number per Month?

```python
avg_deaths_per_month = df.groupby("month").sum()["deaths"]
avg_deaths_per_month
```

```python
avg_deaths_per_month.plot()
```

The line chart depicts a positive linear trend, climaxing in the month of April.

# What are the Top 10 Countries with the Highest Case Number?

```python
sum_cases_per_country = df.groupby("countriesAndTerritories").sum()["cases"]
sum_cases_per_country
```

```python
sum_cases_per_country.nlargest(10).plot.bar()
```

USA has the highest case count globally

# Are the amount of cases related to amount of deaths in Italy?

```python
italy_data = df.loc[df["countriesAndTerritories"] == "Italy"]
italy_data[:10]
```

```python
italy_data.plot.scatter(x="cases", y="deaths")
```

The scatter plot depicts that as the number of cases in Italy increase, the number of deaths also increase.
